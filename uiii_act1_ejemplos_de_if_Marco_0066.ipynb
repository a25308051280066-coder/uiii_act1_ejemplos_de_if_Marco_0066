### Ejemplo de if

```python
# Ejemplo de if simple en Python
# Un ejemplo común es verificar si un usuario es mayor de edad para permitirle el acceso a un servicio:
edad = int(input("Ingresa tu edad: ")) # Solicita el dato al usuario

if edad >= 18: # Evalúa si el valor es mayor o igual a 18
    print("Eres mayor de edad.") # Solo se imprime si la condición se cumple
```

```python
# 2. Ejemplo de if-else con Operadores Relacionales y Aritméticos
# Este código determina si una persona es mayor de edad basándose en el año de nacimiento.
año_actual = 2026 # Asigna el año actual a una variable
año_nacimiento = int(input("Ingresa tu año de nacimiento: ")) # Solicita el año y lo convierte a entero
edad = año_actual - año_nacimiento # Calcula la edad usando el operador aritmético de resta

if edad >= 18: # Evalúa si la edad es mayor o igual a 18 (operador relacional)
    print("Eres mayor de edad.") # Se ejecuta si la condición anterior es verdadera
else: # Se ejecuta si la condición del if fue falsa
    print("Eres menor de edad.") # Indica que la persona tiene menos de 18 años
```

```python
# 3. Ejemplo de if-elif-else con Operadores Relacionales
# Verifica si un número ingresado es positivo, negativo o cero.
numero = int(input("Ingresa un número: ")) # Solicita un número al usuario y lo convierte a entero

if numero > 0: # Compara si el número es mayor a cero
    print("El número es positivo.") # Informa si el número es mayor a cero
elif numero < 0: # Si la primera condición falló, verifica si es menor a cero
    print("El número es negativo.") # Informa si el número es menor a cero
else: # Si no es mayor ni menor a cero, solo queda una opción
    print("El número es cero.") # Informa que el valor es exactamente cero
```

```python
# Solicitar la nota al usuario y convertirla a número decimal
nota = float(input("Ingresa la calificación (0-10): "))

# Primer nivel: Verificar si la nota está dentro del rango permitido (0 a 10)
if nota >= 0 and nota <= 10: # Evalúa el rango válido [5, 6]
    # Segundo nivel: Si la nota es válida, verificar si aprobó o reprobó
    if nota >= 6: # Criterio de aprobación basado en las fuentes [3, 4]
        print("Resultado: Aprobado") # Se ejecuta si la nota es válida y >= 6
    else:
        print("Resultado: Reprobado") # Se ejecuta si la nota es válida pero < 6
else:
    # Este bloque pertenece al primer 'if' y se ejecuta si la nota es menor a 0 o mayor a 10
    print("Error: Nota fuera de rango. Debe estar entre 0 y 10.")
```
